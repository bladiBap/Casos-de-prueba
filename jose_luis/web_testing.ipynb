```python
import os
import random
import unittest
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time as timer
import HtmlTestRunner
from datetime import datetime, timedelta


class WebTesting(unittest.TestCase):
    def setUp(self):
        self.driver = webdriver.Chrome()
        self.driver.get("https://tartesain.com")
        self.driver.maximize_window()
        os.makedirs('img', exist_ok=True)
        self.i = 0

    def get_phone_ramdom(self):
        return random.randint(600000000, 699999999)

    def test_register(self):
        timer.sleep(2)
        # Paso 1: Navegar a la página de inicio de sesión
        auth_link = self.driver.find_element(By.ID, "menu-link-user")
        auth_link.click()
        timer.sleep(2)
        self.assertIn("https://tartesain.com/iniciar-sesion", self.driver.current_url)
        # Paso 2: Navergar a la página de registro
        register_link = self.driver.find_element(By.ID, "link-register")
        register_link.click()
        timer.sleep(2)
        self.assertIn("https://tartesain.com/registro", self.driver.current_url)
        # Paso 3: Ingresar el nombre
        name = self.driver.find_element(By.ID, "input-name")
        name.send_keys("Test")
        timer.sleep(1)
        # Paso 4: Ingresar el apellido
        last_name = self.driver.find_element(By.ID, "input-last-name")
        last_name.send_keys("Test")
        timer.sleep(1)
        # Paso 5: Ingresar el teléfono
        phone = self.driver.find_element(By.ID, "input-phone")
        phone.send_keys(self.get_phone_ramdom().__str__())
        timer.sleep(1)
        # Paso 6: Ingresar el correo
        email = self.driver.find_element(By.ID, "input-email")
        email.send_keys("test"+self.get_phone_ramdom().__str__()+"@gmail.com")
        timer.sleep(1)
        # Paso 7: Ingresar la contraseña
        password = self.driver.find_element(By.ID, "input-password")
        password.send_keys("12345678")
        timer.sleep(1)
        # Paso 8: Confirmar la contraseña
        confirm_password = self.driver.find_element(By.ID, "input-confirm-password")
        confirm_password.send_keys("12345678")
        timer.sleep(1)
        # Paso 9: Dar clic en el botón de registro
        register_button = self.driver.find_element(By.ID, "btn-register")
        register_button.click()
        timer.sleep(1)
        # Paso 10: Verificar que el registro fue exitoso
        toast_container = WebDriverWait(self.driver, 10).until(EC.visibility_of_element_located((By.CLASS_NAME, "toast__container")))
        toast_body = toast_container.find_element(By.CLASS_NAME, "Toastify__toast-body")
        toast_message = toast_body.find_elements(By.TAG_NAME, "div")[1]
        self.assertIn("Registro exitoso, inicia sesión para continuar", toast_message.text)
        timer.sleep(1)


    def test_login(self):
        timer.sleep(2)
        # Paso 1: Navegar a la página de inicio de sesión
        auth_link = self.driver.find_element(By.ID, "menu-link-user")
        auth_link.click()
        timer.sleep(2)
        self.assertIn("https://tartesain.com/iniciar-sesion", self.driver.current_url)
        # Paso 2: Ingresar el correo
        email = self.driver.find_element(By.ID, "input-email")
        email.send_keys("test@gmail.com")
        timer.sleep(1)
        # Paso 3: Ingresar la contraseña
        password = self.driver.find_element(By.ID, "input-password")
        password.send_keys("12345678")
        timer.sleep(1)
        # Paso 4: Dar clic en el botón de inicio de sesión
        phone = self.driver.find_element(By.ID, "btn-login")
        phone.click()
        timer.sleep(2)
        self.assertIn("https://tartesain.com/perfil", self.driver.current_url)
        timer.sleep(1)


    def test_add_product_to_cart(self):
        timer.sleep(2)
        # Paso 1: Dar clic en el botón de productos
        products_button = self.driver.find_element(By.ID, "menu-btn-products")
        products_button.click()
        timer.sleep(2)
        # Paso 2: Navegar a la página de productos
        products_link = self.driver.find_element(By.ID, "menu-link-all-products")
        products_link.click()
        timer.sleep(2)
        self.assertIn("https://tartesain.com/productos", self.driver.current_url)
        # Paso 3: Dar clic en el botón de agregar al carrito del primer producto
        add_to_cart_button = self.driver.find_element(By.CSS_SELECTOR, "[id^='btn-add-to-cart-']")
        add_to_cart_button.click()
        timer.sleep(2)
        # Paso 4: Navergar al carrito de compras
        cart_link = self.driver.find_element(By.ID, "menu-link-cart")
        cart_link.click()
        timer.sleep(2)
        self.assertIn("https://tartesain.com/carrito", self.driver.current_url)
        # Paso 5: Verificar que el producto fue agregado al carrito
        product_container = self.driver.find_element(By.CSS_SELECTOR, "[class^='product_container']")
        self.assertIsNotNone(product_container)
        timer.sleep(1)


    def test_remove_product_from_cart(self):
        timer.sleep(2)
        # Paso 1: Dar clic en el botón de productos
        products_button = self.driver.find_element(By.ID, "menu-btn-products")
        products_button.click()
        timer.sleep(2)
        # Paso 2: Navegar a la página de productos
        products_link = self.driver.find_element(By.ID, "menu-link-all-products")
        products_link.click()
        timer.sleep(2)
        self.assertIn("https://tartesain.com/productos", self.driver.current_url)
        # Paso 3: Dar clic en el botón de agregar al carrito del primer producto
        add_to_cart_button = self.driver.find_element(By.CSS_SELECTOR, "[id^='btn-add-to-cart-']")
        add_to_cart_button.click()
        timer.sleep(2)
        # Paso 4: Navergar al carrito de compras
        cart_link = self.driver.find_element(By.ID, "menu-link-cart")
        cart_link.click()
        timer.sleep(2)
        self.assertIn("https://tartesain.com/carrito", self.driver.current_url)
        # Paso 5: Verificar que el producto fue agregado al carrito
        product_container = self.driver.find_element(By.CSS_SELECTOR, "[class^='product_container']")
        self.assertIsNotNone(product_container)
        timer.sleep(1)
        # Paso 6: Dar clic en el botón de eliminar del primer producto
        remove_button = product_container.find_element(By.CSS_SELECTOR, ".btn-remove-item")
        remove_button.click()
        timer.sleep(2)
        # Paso 7: Verificar que el producto fue eliminado del carrito
        empty_cart_message = self.driver.find_element(By.CSS_SELECTOR, "h3")
        self.assertEqual(empty_cart_message.text, "Tu carrito de compras está vacío")
        timer.sleep(1)


    def test_checkout(self):
        timer.sleep(2)
        # Paso 1: Dar clic en el botón de productos
        products_button = self.driver.find_element(By.ID, "menu-btn-products")
        products_button.click()
        timer.sleep(2)
        # Paso 2: Navegar a la página de productos
        products_link = self.driver.find_element(By.ID, "menu-link-all-products")
        products_link.click()
        timer.sleep(2)
        self.assertIn("https://tartesain.com/productos", self.driver.current_url)
        # Paso 3: Dar clic en el botón de agregar al carrito del primer producto
        add_to_cart_button = self.driver.find_element(By.CSS_SELECTOR, "[id^='btn-add-to-cart-']")
        add_to_cart_button.click()
        timer.sleep(2)
        # Paso 4: Navergar al carrito de compras
        cart_link = self.driver.find_element(By.ID, "menu-link-cart")
        cart_link.click()
        timer.sleep(2)
        self.assertIn("https://tartesain.com/carrito", self.driver.current_url)
        # Paso 5: Verificar que el producto fue agregado al carrito
        product_container = self.driver.find_element(By.CSS_SELECTOR, "[class^='product_container']")
        self.assertIsNotNone(product_container)
        timer.sleep(1)
        # Paso 6: Dar clic en el botón de finalizar compra
        checkout_button = self.driver.find_element(By.ID, "btn-checkout")
        checkout_button.click()
        timer.sleep(2)
        # Paso 7: Verificar que se redirigió a la página de inicio de sesión
        self.assertIn("https://tartesain.com/iniciar-sesion", self.driver.current_url)
        timer.sleep(1)
        # Paso 8: Ingresar credenciales de inicio de sesión
        email_input = self.driver.find_element(By.ID, "input-email")
        email_input.send_keys("test@gmail.com")
        timer.sleep(1)
        password_input = self.driver.find_element(By.ID, "input-password")
        password_input.send_keys("12345678")
        timer.sleep(1)
        # Paso 9: Dar clic en el botón de iniciar sesión
        login_button = self.driver.find_element(By.ID, "btn-login")
        login_button.click()
        # Paso 10: Verificar que se redirigió a la página de finalizar compra
        timer.sleep(2)
        self.assertIn("https://tartesain.com/finalizar-compra", self.driver.current_url)
        # Paso 11: Ingresar dirección de envío
        address_input = self.driver.find_element(By.ID, "input-address")
        address_input.send_keys("Universidad NUR, Av. Cristo Redentor")
        timer.sleep(1)# Paso 12: Seleccionar la ubicación de envío
        location_select = self.driver.find_element(By.ID, "select-ubication")
        location_select.click()
        timer.sleep(1)
        location_option = self.driver.find_element(By.CSS_SELECTOR, "[id='4-anillo']")
        location_option.click()
        timer.sleep(1)
        # Paso 13: Seleccionar la fecha de entrega        
        # Calcula la fecha objetivo sumando 5 días a la fecha actual
        target_date = datetime.now() + timedelta(days=5)
        # Verifica si la fecha objetivo cae en un fin de semana
        if target_date.weekday() == 5:  # Si es sábado (5)
            target_date += timedelta(days=2)  # Avanza al lunes
        elif target_date.weekday() == 6:  # Si es domingo (6)
            target_date += timedelta(days=1)  # Avanza al lunes
        day = target_date.day
        month = target_date.month
        year = target_date.year
        date_input = self.driver.find_element(By.ID, "fecha-entrega")
        date_input.click()
        day_input = WebDriverWait(self.driver, 10).until(
            EC.element_to_be_clickable((By.XPATH, "//div[@data-type='day']"))
        )
        day_input.click()
        day_input.send_keys(day)
        timer.sleep(1)
        month_input = WebDriverWait(self.driver, 10).until(
            EC.element_to_be_clickable((By.XPATH, "//div[@data-type='month']"))
        )
        month_input.click()
        month_input.send_keys(month)
        timer.sleep(1)
        year_input = WebDriverWait(self.driver, 10).until(
            EC.element_to_be_clickable((By.XPATH, "//div[@data-type='year']"))
        )
        year_input.click()
        year_input.send_keys(year)
        timer.sleep(1)
        # Paso 14: Seleccionar la hora de entrega
        time_input = self.driver.find_element(By.ID, "hora-entrega")
        time_input.click()
        hour_input = WebDriverWait(self.driver, 10).until(
            EC.element_to_be_clickable((By.XPATH, "//div[@data-type='hour']"))
        )
        hour_input.click()
        hour_input.send_keys("07")
        timer.sleep(1)
        minute_input = WebDriverWait(self.driver, 10).until(
            EC.element_to_be_clickable((By.XPATH, "//div[@data-type='minute']"))
        )
        minute_input.click()
        minute_input.send_keys("30")
        timer.sleep(1)
        # Paso 15: Dar clic en el botón de finalizar compra
        finish_button = self.driver.find_element(By.ID, "btn-finalizar-compra")
        finish_button.click()
        timer.sleep(2)
        # Paso 16: Verificar que el pedido fue realizado
        toast_container = WebDriverWait(self.driver, 10).until(EC.visibility_of_element_located((By.CLASS_NAME, "toast__container")))
        toast_body = toast_container.find_element(By.CLASS_NAME, "Toastify__toast-body")
        toast_message = toast_body.find_elements(By.TAG_NAME, "div")[1]
        self.assertIn("Compra realizada con éxito", toast_message.text)
        timer.sleep(1)


    def test_contact_form(self):
        timer.sleep(2)
        # Paso 1: Navegar a la página de contacto
        contact_link = self.driver.find_element(By.ID, "menu-link-contact")
        contact_link.click()
        timer.sleep(2)
        self.assertIn("https://tartesain.com/contacto", self.driver.current_url)
        # Paso 2: Ingresar el nombre completo
        full_name = self.driver.find_element(By.ID, "input-fullname")
        full_name.send_keys("John Doe")
        timer.sleep(1)
        # Paso 3: Ingresar el correo electrónico
        email = self.driver.find_element(By.ID, "input-email")
        email.send_keys("test@gmail.com")
        timer.sleep(1)
        # Paso 4: Ingresar el número de teléfono
        phone = self.driver.find_element(By.ID, "input-phone")
        phone.send_keys("1234567890")
        timer.sleep(1)
        # Paso 5: Ingresar el asunto
        subject = self.driver.find_element(By.ID, "input-subject")
        subject.send_keys("Test")
        timer.sleep(1)
        # Paso 6: Ingresar el mensaje
        message = self.driver.find_element(By.ID, "input-message")
        message.send_keys("Test message")
        timer.sleep(1)
        # Paso 7: Enviar el formulario
        submit_button = self.driver.find_element(By.ID, "btn-send-email")
        submit_button.click()
        timer.sleep(3)
        # Paso 8: Validar mensaje de éxito
        toast_container = WebDriverWait(self.driver, 10).until(EC.visibility_of_element_located((By.CLASS_NAME, "toast__container")))
        toast_body = toast_container.find_element(By.CLASS_NAME, "Toastify__toast-body")
        toast_message = toast_body.find_elements(By.TAG_NAME, "div")[1]
        self.assertIn("Mensaje enviado correctamente", toast_message.text)
        timer.sleep(1)


    def tearDown(self):
        self.driver.close()


if __name__ == "__main__":
    with open("report.html", "w") as file:
        runner = HtmlTestRunner.HTMLTestRunner(output='reports', verbosity=2)
        runner.run(unittest.TestLoader().loadTestsFromTestCase(WebTesting))
```
